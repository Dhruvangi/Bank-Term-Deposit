 # How did you Pick best model using accuracy precision and recall?

Here, I consider Decision Tree with all variables as best model for given Problem.
- AS, accuracy is 0.81 that means 81 % of the time model is  correctly classifying   the "Yes" class(potential clients that are willing to subcribe a term deposit) and "No" class(potential clients that are not willing to suscribe a term deposit). 
- The recall is the measure of our model correctly identifying True Positives. 
   - In given case we dont want to loose clinents who would have subscribe for term deposit, but our model predicted negatively. In this situation our Recall Should be high(low false negative) which helps to minimize false prediction.
-  precision indicates the rate at which positive predictions are correct.
  - In other words, it minimize the incorrect classification of potential clients that are not willing to subcribe a term     deposit but model predicted as positive. 
  - Which ultimately help to save time and cost that we are going to invest on clients to educate about bank term deposit       who are not going to subscribe term deposit but model predicted as positive.
